# Predict by passing image path

```python

# Disable scientific notation for clarity
np.set_printoptions(suppress=True)

# Load the model
model = tensorflow.keras.models.load_model('model3.h5')

# Create the array of the right shape to feed into the keras model
# The 'length' or number of images you can put into the array is
# determined by the first position in the shape tuple, in this case 1.
data = np.ndarray(shape=(1, 224, 224, 3), dtype=np.float32)

# Replace this with the path to your image
image = Image.open(r'C:\Users\Hitesh khatana\Desktop\projects\with or without mask\masks2.0\masks\test\14.jpeg')

#resize the image to a 224x224 with the same strategy as in TM2:
#resizing the image to be at least 224x224 and then cropping from the center
size = (224, 224)
image = ImageOps.fit(image, size, Image.ANTIALIAS)

#turn the image into a numpy array
image_array = np.asarray(image)

# display the resized image
image.show()

# Normalize the image
normalized_image_array = (image_array.astype(np.float32) / 127.0) - 1

# Load the image into the array
data[0] = normalized_image_array

# run the inference
prediction = model.predict(data)

if prediction[0][0]>prediction[0][1]:
    print("person is wearing mask")
    
else:
    print("Person is not wearing mask")
    
```

# Predict using webcam

```python
model = tensorflow.keras.models.load_model('model3.h5')

data = np.ndarray(shape=(1,480,640, 3), dtype=np.float32)

face_cascade = cv2.CascadeClassifier("haarcascade.xml")
```

```python
video = cv2.VideoCapture(0)
ch = True
while ch == True:
    c ,frame = video.read()
    
    faces = face_cascade.detectMultiScale(frame , scaleFactor = 1.05 , minNeighbors = 5)
    
    for x,y,w,h in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)
        font = cv2.FONT_HERSHEY_PLAIN         
    
    np.set_printoptions(suppress=True)
    #img = cv2.resize(frame , (224 , 224 )) 
    normalized_image_array = (frame.astype(np.float32) / 127.0) - 1


    data[0] = normalized_image_array

    prediction = model.predict(data)

    if prediction[0][0] > prediction[0][1]:
        cv2.putText(frame ,"person is wearing mask" , (x,y-20) , font , 1 , (255,255,255))
    #if prediction[0][1] > prediction[0][0] and prediction[0][1] > prediction[0][2]:
    #    cv2.putText(frame ,"person is not wearing mask properly " , (x,y-20) , font , 1 , (255,255,255))
    if prediction[0][1] > prediction[0][0]:
        cv2.putText(frame ,"person is not wearing mask" , (x,y-20) , font , 1 , (255,255,255))  
    cv2.imshow("capturing", frame)
    key = cv2.waitKey(1)
    if key == ord('c'):
        ch = False
        img = frame
        video.release()
        cv2.destroyAllWindows()
    
```

```python
factors = data.drop(["result" , "Unnamed: 0" ], axis = 1)
target = data["result"]
```

```python
model = tensorflow.keras.models.load_model('model3.h5')
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(factors,target,test_size=.2, random_state = 4)
```

```python
model.compile(loss='categorical_crossentropy', optimizer="adam", metrics=['accuracy'])
```

```python
model.fit(x_train,y_train, validation_data=(x_test, y_test), epochs=10)
```
